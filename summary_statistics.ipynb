```python
import pandas as pd
```

```python
# Read CSV file
csv = pd.read_csv("csv/data11-21.csv")
df = pd.DataFrame(csv)
```

```python
# Count length of data
data_size = df.shape[0]

normal_fault = df[df["断層型"] == "正"]
reverse_fault = df[df["断層型"] == "逆"]
right_sideway = df[df["断層型"] == "右横ずれ"]
left_sideway = df[df["断層型"] == "左横ずれ"]
sideway = df[df["断層型"] == "横ずれ"]
```

```python
 # Test matching data_size
sum_fault = normal_fault.shape[0] + reverse_fault.shape[0] + right_sideway.shape[0] + left_sideway.shape[0] + sideway.shape[0]
print(sum_fault == data_size)

vertical = pd.concat([normal_fault, reverse_fault])
sideway = pd.concat([right_sideway, left_sideway, sideway])
```

```python
# Test matching data_size
sum_fault = vertical.shape[0] + sideway.shape[0]
print(sum_fault == data_size)
```

```python
# Get displacement speed
sideway_speed = sideway[["名前", "平均変位速度(調整後)[m/千年]"]]
vertical_speed = vertical[["名前", "平均変位速度(調整後)[m/千年]"]] 
vertical_speed = vertical_speed.reset_index(drop=True)
sideway_speed = sideway_speed.reset_index(drop=True)

```

```python
sideway_speed = sideway_speed[sideway_speed["平均変位速度(調整後)[m/千年]"] != "-"]
sideway_speed["平均変位速度(調整後)[m/千年]"] = sideway_speed["平均変位速度(調整後)[m/千年]"].astype(str).astype(float)
sideway_speed = sideway_speed.rename(columns={"平均変位速度(調整後)[m/千年]": 'displacement speed (m/year)'})
sideway_speed = sideway_speed.sort_values(by="displacement speed (m/year)")
sideway_speed
```

```python

sideway_speed.to_csv("csv/sideway_speed.csv")
```

```python
vertical_speed = vertical_speed[vertical_speed["平均変位速度(調整後)[m/千年]"] != "-"]
vertical_speed["平均変位速度(調整後)[m/千年]"] = vertical_speed["平均変位速度(調整後)[m/千年]"].astype(str).astype(float)
vertical_speed = vertical_speed.rename(columns = {"平均変位速度(調整後)[m/千年]": 'displacement speed (m/year)'})
vertical_speed = vertical_speed.sort_values(by="displacement speed (m/year)")
vertical_speed
```

```python

vertical_speed.to_csv("csv/vertical_speed.csv")
```
